# Morphing

```
trial = cv2.erode(sh_A,se_B1,iterations = 1)
trial2 = cv2.dilate(sh_A,se_B1,iterations = 1)
```

```
fig = plt.figure(figsize = [20,20])
ax1 = fig.add_subplot(1,3,1)
ax1.set_title("Original Shape", fontsize = 16)
ax1.imshow(sh_A, cmap = 'gray')
ax2 = fig.add_subplot(1,3,2)
ax2.set_title("Eroded Shape", fontsize =16)
ax2.imshow(trial, cmap = 'gray')
ax3 = fig.add_subplot(1,3,3)
ax3.set_title("Dilated Shape", fontsize = 16)
ax3.imshow(trial2, cmap = 'gray')
fig.suptitle('', fontsize=20)
```

```
def morph(img,se,label1,label2):
  #se_A
  trial = cv2.erode(img,se,iterations = 1)
  trial2 = cv2.dilate(img,se,iterations = 1)
  
  fig = plt.figure(figsize = [10,10])
  fig.suptitle(str(label1) + " and " + str(label2), fontsize=20, ha='center', va = 'bottom')
  ax1 = fig.add_subplot(1,3,1)
  ax1.set_title("Original Shape", fontsize = 16)
  ax1.imshow(img, cmap = 'gray')
  ax2 = fig.add_subplot(1,3,2)
  ax2.set_title("Eroded Shape", fontsize =16)
  ax2.imshow(trial, cmap = 'gray')
  ax3 = fig.add_subplot(1,3,3)
  ax3.set_title("Dilated Shape", fontsize = 16)
  ax3.imshow(trial2, cmap = 'gray')
```

```
images = [sh_A, sh_B, sh_C, sh_D]
SEs = [se_A , se_B1, se_B2, se_C1, se_C2, se_D, se_E1, se_E2]
images_label = ["Square", "Triangle", "Hollow", "Cross"]
SEs_label = ["2x2" , "2x1a", "2x1b", "1x2a", "1x2b", "cross", "diagonalA", "diagonalB"]
```

```
icount = 0
for i in images:
  jcount = 0
  for j in SEs:
    ilabel = images_label[icount]
    jlabel = SEs_label[jcount]
    morph(i,j,ilabel,jlabel)
    jcount+=1
  icount+=1  
```

```
fig = plt.figure(figsize = [20,20])
ax1 = fig.add_subplot(1,5,1)
ax1.set_title("2x2", fontsize =16)
ax1.imshow(se_A, cmap = 'gray')
ax2 = fig.add_subplot(1,5,2)
ax2.set_title("2x1", fontsize = 16)
ax2.imshow(se_B2, cmap = 'gray')
ax3 = fig.add_subplot(1,5,3)
ax3.set_title("1x2", fontsize = 16)
ax3.imshow(se_C1, cmap = 'gray')
ax4 = fig.add_subplot(1,5,4)
ax4.set_title("Cross", fontsize = 16)
ax4.imshow(se_D, cmap = 'gray')
ax5 = fig.add_subplot(1,5,5)
ax5.set_title("Diagonal", fontsize = 16)
ax5.imshow(se_E1, cmap = 'gray')
```
